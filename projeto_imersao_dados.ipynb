```python
# seaborn correlation matrix

import numpy as np

# Compute the correlation matrix
corr = dados.loc[:, 'g-0': 'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Compute the correlation matrix
corr_celular = dados.loc[:, 'c-0': 'c-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

#não podemos inferir causalidade a partir da correlação
```

# Aula 4 - Merge de dados e análise de resultados

```python
# nessa tabela temos os dados de quais mecanismos de açao especificamente foram ativados
# as colunas sao os mecanismos e as linhas tem os ids dos experimentos
# verificando a tabela podemos notar que alguns experimentos ativam nenhum outro ativam 1 ou mais mecanismos

dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados.head()
```

```python
# verificando os dados das colunas/mecanismos de açao
#id é a droga
# os valores sao 1 ou 0 e significam respctivamente que o mecanismo de açao foi ativado e nao ativado
dados_resultados['adenylyl_cyclase_activator'].unique()
```

```python
## verificar o mecanismo de açao mais ativado por experimento

##para isso primeiro verificar se as colunas possuem somente números ou se possuem strings também
dados_resultados.info()

## ha duas formas para encontrar o maior numero de mecanismos ativos
# contagem_moa = dados_resultados.drop('id', axis=1).sum().sort_values(ascending=False)
# contagem_moa
contagem_moa = dados_resultados.select_dtypes('int64').sum()
contagem_moa.sort_values(ascending=False)
```

```python
# soma da linha para ver quantos mecanismos de ativaçao foram ativados em cada linha/droga usaada
# dados_resultados.sum()
dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
# criando novas colunas
# cria uma coluna no fim da tabela 'n_moa'
# essa coluna vai ser o numero de mecanismos ativos por droga calculado a cima 

dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
# juntar as tabelas com merge do pandas

dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa', 'ativo_moa']], on='id')
dados_combinados.head()
```

# Aula 5 - Machine Learning, Sci-kit learning e desafios envolvidos

```python
# cada linha da tabela original tem uma assinatura
# em cada experimento teve suas caracteristicas de tempo, composto e tratamento 
dados.head()
# com machine learning vamos dar ao algoritimo uma nova assinatura
# e ele tera que prever se aconteceu alguma ativaçao de mecanismo
#como um composto pode acionar mais de um mecanismo, nós teremos uma classificação multilabel

dados_resultados.head()
#dados_resultados é como se fosse a resposta da primeira
#foi construída por cientistas que analisaram o efeito de um determinado composto nos mais de 700 genes
```

```python
# from sklearn.model_selection import train_test_split
# y_teste = train_test_split(x, y, test_size=0.2)
```

```python
# queremos que o modelo faça predições compatíveis com a última coluna da nossa base de dados dados_combinados a ativo_moa.
#vamos usar a Regressão Logística

# import da biblioteca 
from sklearn.linear_model import LogisticRegression

# funçao que separa parte dos dados para usar como teste do modelo
from sklearn.model_selection import train_test_split

# dados com assinatura celular, sao os com type float64
x = dados_combinados.select_dtypes('float64')

# respostas do modelo
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, stratify=y, random_state=234)
#20% dos dados serao usados pra teste, o resto sera pro treino
#o parametro random_state impede que a cada vez que o algoritimo rodar tera um resultado diferente 
#o parametro stratify garante a separaçao com proporçao correta dos dados pros testes
#garantindo que o resultado do modelo simples seja igual a quantidade de mecanismos ativos
#ja que ele chutou que foram ativos em todos experimentos

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino) # metodo fit recebe os parametros de treino e faz o treinamento do modelo
modelo_rlogistica.score(x_teste, y_teste) # recebe os parametros de teste e verifica a precisao do modelo
#para ter noçao se a precisao do modelo foi alta podemos comparar com o resultado de algum modelo mais simples
```

```python
# importando modulo pra fazer o modelo simples
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

# estrategia que o modelo ira seguir
# ele vai analisar os dados e identificar o resultado mais frequente do mecanismo de ativaçao
# e chutar que todos os dados de testes serao iguais
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino) # treino do modelo

previsao_dummy = modelo_dummy.predict(x_teste) #prediçao do resultado
accuracy_score(y_teste, previsao_dummy) #calcula a acuracia definitiva
```

```python
# verificando ocorrencia de true e false para comparar com o resultado do modelo
dados_combinados['ativo_moa'].value_counts(normalize=True)
```

```python
# usando outro modelo

from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, stratify=y, random_state=234)

modelo_arvore = DecisionTreeClassifier(max_depth = 3) #profundidade da árvore
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(15, 10),facecolor='k')
tree.plot_tree(
    modelo_arvore,
    ax=ax,
    fontsize=10,
    rounded=True,
    filled=True,
    feature_names=x_treino.columns,
    class_names=['Não ativado', 'Ativado']
)
plt.show()

#a arvore é criada a partir de um no raiz que é uma var
#aumentando o numero de niveis da arvore alcançamos uma precisao maior no modelo
#O último nível da árvore, apresenta n quadros 
#que apresentam características que devem ser levadas em conta na hora da classificação
#Esse nível não apresenta mais ramificações e, esses quadros são chamados de folhas.
```

```python
#como o resultado com 3 niveis nao foi satisfatorio, vamos alterar e executar o codigo com um laçode repetiçao

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, stratify=y, random_state=234)

#variaveis criadas pra guardar o resultado do modelo em cada iteraçao para compara-las
teste=[]
treino=[]

#serao criadas 15 arvores
for i in range(1, 10):
  modelo_arvore = DecisionTreeClassifier(max_depth = i)
  modelo_arvore.fit(x_treino, y_treino)
  teste.append( modelo_arvore.score(x_teste, y_teste) )
  treino.append( modelo_arvore.score(x_treino, y_treino) )
```

```python
#plot dos resultados, pode se notar que treino evolui bem 
sns.lineplot(x=range(1,10), y=teste, label='teste')
sns.lineplot(x=range(1,10), y=treino, label='treino')
#ocorre overfitting

#conforme aumentamos a profundidade da nossa árvore de decisão, a classificação para os dados de treino ficam muito boas
#pois ela consegue captar muito bem as características desses dados
#Mas, o modelo fica tão bom os dados de treino que, quando apresentamos um novo conjunto de dados (teste), ele não consegue generalizar tão bem 
#percebemos que, para melhorar o nosso modelo, não é suficiente somente aumentar a profundidade da árvore de decisão
```
